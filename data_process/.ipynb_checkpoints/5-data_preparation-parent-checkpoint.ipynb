## Step 1. 准备全局 GGI 网络
确定所有 gene 节点的全集（两列，以空格分割）
___

```python
import os
from tqdm import tqdm
from pathlib import Path
```

```python
# ===== 根目录路径 =====
DATA_ROOT = Path("/apdcephfs_qy3/share_2932069/ziyiyang/1-antibiotic_resistance/PANACEA/data/3-string_ppi_data")
OUTPUT_ROOT = Path("/opt/ai4g_chriszyyang/buddy1/2_project_ongoing/4-antibio_resistance/PANACEA/data/5-network/")

# 输出文件
OUT_FILE = OUTPUT_ROOT / "global_ggi_edgelist.txt"
```

```python
edges = set()

files = list(DATA_ROOT.rglob("*.ggi.in_genes.tsv"))
print(f"Found {len(files)} files.")

for tsv_path in tqdm(files):
    with tsv_path.open("r", encoding="utf-8", errors="ignore") as f:
        # ---- 跳过表头 ----
        header = next(f, None)
        
        # ---- 逐行读取数据 ----
        for line in f:
            line = line.strip()
            if not line:
                continue
            parts = line.split("\t")
            if len(parts) < 2:
                continue

            g1, g2 = parts[0].strip(), parts[1].strip()

            if not g1 or not g2 or g1 == g2:
                continue

            # 无向边
            if g1 < g2:
                edge = (g1, g2)
            else:
                edge = (g2, g1)

            edges.add(edge)

len(edges)
```

```python
with OUT_FILE.open("w", encoding="utf-8") as out:
    for g1, g2 in sorted(edges):
        out.write(f"{g1} {g2}\n")

print(f"Saved global edgelist to:\n{OUT_FILE}")
```

## Step 2. 准备 Meta-graph 网络
构建 GO annotation - Strains 的层级关系
___

```python
from pathlib import Path

GO_ROOT = Path("/opt/ai4g_chriszyyang/buddy1/2_project_ongoing/4-antibio_resistance/PANACEA/data/4-go_annotation")
OUT_FILE = Path("/opt/ai4g_chriszyyang/buddy1/2_project_ongoing/4-antibio_resistance/PANACEA/data/5-network/mg_edgelist.txt")

OUT_FILE.parent.mkdir(parents=True, exist_ok=True)
```

```python
from tqdm import tqdm
import itertools

edges = set()             # (nodeA, nodeB)
species_to_strains = {}   # species → list of strains

strain_dirs = [p for p in GO_ROOT.iterdir() if p.is_dir()]
print(f"Found {len(strain_dirs)} strain folders.")

def get_species_from_folder(folder_name):
    """species 是 去掉最后一个 '_' 之后的部分"""
    if "_" not in folder_name:
        return folder_name
    return "_".join(folder_name.split("_")[:-1])

for strain_dir in tqdm(strain_dirs):
    strain_name = strain_dir.name
    species = get_species_from_folder(strain_name)

    # 收集 species → strain
    species_to_strains.setdefault(species, []).append(strain_name)

    # GO edges
    for tsv_path in strain_dir.glob("*.go_raw.tsv"):
        with tsv_path.open("r", encoding="utf-8", errors="ignore") as f:
            next(f, None)  # skip header

            for line in f:
                parts = line.strip().split("\t")
                if len(parts) < 5:
                    continue

                go_id = parts[3].strip()
                go_type = parts[4].strip()

                if not go_id or not go_type:
                    continue

                go_node = f"{go_id}_{go_type}"

                edges.add((go_node, strain_name))  # GO → strain
```

```python
strain_edges = set()

for species, strain_list in species_to_strains.items():
    if len(strain_list) < 2:
        continue

    # 对每个 species 内的 strain 做两两组合
    for a, b in itertools.combinations(strain_list, 2):
        # 无向边归一化
        edge = tuple(sorted((a, b)))
        strain_edges.add(edge)

len(strain_edges)
```

```python
# 统一合并：GO-strain edges + strain-strain edges
final_edges = set()

# 添加 GO → strain edges
for a, b in edges:
    final_edges.add((a, b))

# 添加 strain ↔ strain edges
for a, b in strain_edges:
    final_edges.add((a, b))

print("Total edges:", len(final_edges))

# 写出成 tsv 格式
with OUT_FILE.open("w", encoding="utf-8") as out:
    for a, b in sorted(final_edges):
        out.write(f"{a}\t{b}\n")

print("Saved to:", OUT_FILE)
```

## Step 2. GO annotation for upstream parents 3 level and SLIM
___

```python
# %% [markdown]
# # 构建 GO annotation - strains meta-graph 的 mg_edgelist
# 
# 功能：
# 1. 解析 GO 本体（go-basic.obo，含 goslim_generic + goslim_prok）
# 2. 对每个 GO_ID 做 DAG 向上 3 层扩展（is_a / part_of）
# 3. 在祖先中筛选落在 GO slim（generic + prok）里的 term
# 4. 为每个 strain 构建：
#    - GO_slim_term <-> strain 的边
#    - 同一 species 下的 strain <-> strain 的边
# 5. 输出统一的 mg_edgelist.txt（两列，用 \t 分隔）
#
# 目录结构假定：
# /opt/.../PANACEA/data/
#   ├─ 4-go_annotation/
#   │    └─ <strain_dir>/
#   │         └─ <strain_dir>.go_raw.tsv   # 含 Genes, TaxID, UniProtID, GO_ID, GO_Type
#   └─ 5-network/
#         └─ mg_edgelist.txt               # 输出

# %% [code]
import os
import re
from collections import defaultdict, deque

import pandas as pd

# ==========================
# 配置区（按需修改）
# ==========================

DATA_ROOT = "/opt/ai4g_chriszyyang/buddy1/2_project_ongoing/4-antibio_resistance/PANACEA/data"
GO_DIR = os.path.join(DATA_ROOT, "4-go_annotation")
NETWORK_DIR = os.path.join(DATA_ROOT, "5-network")
os.makedirs(NETWORK_DIR, exist_ok=True)

MG_EDGE_FILE = os.path.join(NETWORK_DIR, "mg_edgelist.txt")

# GO 本体文件
GO_OBO_FILE = f"{DATA_ROOT}/1-processed_data/go-basic.obo"

# upward propagation 最大深度
UPWARD_DEPTH = 3

# 需要使用的 GO slim subset 名称（generic + prok）
SLIM_SUBSETS = {"goslim_prok", "goslim_generic"}
```

```python
# %% [markdown]
# ## 解析 GO 本体（go-basic.obo）
# 提取：
# - parents: 每个 GO term 的父节点（is_a / part_of）
# - name_dict: GO term -> 名称
# - namespace_dict: GO term -> namespace (BP/MF/CC)
# - subset_dict: GO term -> subset 标签集合（用于识别 GO slim）

# %% [code]
def parse_go_obo(obo_file):
    """
    解析 go-basic.obo，提取：
    - parents: 每个 GO term 的父节点（仅 is_a / part_of）
    - name_dict: GO term -> name
    - namespace_dict: GO term -> namespace (biological_process / molecular_function / cellular_component)
    - subset_dict: GO term -> set of subset 标签
    """
    parents = defaultdict(set)
    name_dict = {}
    namespace_dict = {}
    subset_dict = defaultdict(set)

    with open(obo_file, "r", encoding="utf-8") as f:
        current_id = None
        for line in f:
            line = line.strip()
            if not line:
                continue

            if line == "[Term]":
                current_id = None
                continue
            if line == "[Typedef]":
                # 后面是 typedef，直接跳过
                current_id = None
                continue

            if line.startswith("id: GO:"):
                current_id = line.split("id: ")[1].strip()
            elif current_id is not None:
                if line.startswith("name:"):
                    name = line.split("name: ", 1)[1].strip()
                    name_dict[current_id] = name
                elif line.startswith("namespace:"):
                    ns = line.split("namespace: ", 1)[1].strip()
                    namespace_dict[current_id] = ns
                elif line.startswith("is_a:"):
                    # is_a: GO:0005886 ! plasma membrane
                    parent_id = line.split("is_a: ")[1].split(" ! ")[0].strip()
                    parents[current_id].add(parent_id)
                elif line.startswith("relationship:"):
                    # 只保留 part_of 关系作为父节点（可选）
                    # e.g., relationship: part_of GO:0005886 ! plasma membrane
                    parts = line.split()
                    if len(parts) >= 3 and parts[1] == "part_of":
                        parent_id = parts[2].strip()
                        parents[current_id].add(parent_id)
                elif line.startswith("subset:"):
                    subset = line.split("subset: ")[1].strip()
                    subset_dict[current_id].add(subset)

    return parents, name_dict, namespace_dict, subset_dict
```

```python
# %% [markdown]
# ## 向上 3 层找祖先 + 构建 GO slim term 集合

# %% [code]
from collections import deque

def get_ancestors(go_id, parents, max_depth=3):
    """
    从给定 GO term 出发，沿着 parents（is_a / part_of）向上
    进行 BFS，返回包含自身在内的所有祖先（深度 <= max_depth）
    """
    visited = set()
    queue = deque()
    queue.append((go_id, 0))
    visited.add(go_id)

    while queue:
        cur, depth = queue.popleft()
        if depth >= max_depth:
            continue
        for p in parents.get(cur, []):
            if p not in visited:
                visited.add(p)
                queue.append((p, depth + 1))

    return visited


def build_slim_term_set(subset_dict, slim_subsets):
    """
    从 subset_dict 中筛出属于 slim_subsets 的 GO term
    """
    slim_terms = set()
    for go_id, subsets in subset_dict.items():
        if subsets & slim_subsets:
            slim_terms.add(go_id)
    return slim_terms
```

```python
# %% [markdown]
# ## namespace -> C/P/F 前缀 & GO label 统一格式
# 
# 输出的 GO 节点格式为：
# `GO:0005886_C:plasma membrane`

# %% [code]
def ns_to_prefix(ns):
    """
    将 GO namespace 映射为 C/P/F 前缀
    """
    if ns == "cellular_component":
        return "C"
    elif ns == "biological_process":
        return "P"
    elif ns == "molecular_function":
        return "F"
    else:
        # 未知时保留原 ns 的首字母
        return ns[:1].upper() if ns else "U"


def format_go_label(go_id, name_dict, namespace_dict):
    """
    将 GO term 格式化成：
    GO:0005886_C:plasma membrane
    """
    name = name_dict.get(go_id, "").replace("\t", " ").strip()
    ns = namespace_dict.get(go_id, "")
    prefix = ns_to_prefix(ns)
    return f"{go_id}_{prefix}:{name}" if name else f"{go_id}_{prefix}"
```

```python
# %% [markdown]
# ## 从 strain 目录名解析 species 名
# 
# 规则示例：
# - `Enterococcus_faecalis_AR-0780`  → `Enterococcus faecalis`
# - `Escherichia_coli_339942-001-501` → `Escherichia coli`

# %% [code]
def parse_species_from_strain_dir(strain_dir_name):
    """
    例如：
    Enterococcus_faecalis_AR-0780          -> Enterococcus faecalis
    Escherichia_coli_339942-001-501       -> Escherichia coli

    简单规则：取前两个下划线分隔的 token 作为 species
    """
    tokens = strain_dir_name.split("_")
    if len(tokens) >= 2:
        return f"{tokens[0]} {tokens[1]}"
    elif len(tokens) == 1:
        return tokens[0]
    else:
        return strain_dir_name
```

```python
# %% [markdown]
# ## 主流程：构建 GO_slim–strain & strain–strain 边并写出 mg_edgelist
# 
# 1. 解析 OBO，构建 parents / name / namespace / subset
# 2. 构建 slim_terms（generic + prok）
# 3. 遍历 4-go_annotation 下的每个 strain 目录：
#    - 读取 `<strain>.go_raw.tsv`
#    - 提取 unique GO_ID
#    - 向上 3 层找祖先 → 取落在 slim 里的 term（若无，则退回原始 GO）
#    - 生成 GO_label → strain 的边
# 4. 对同一 species 下的所有 strain 构建完全图的 strain–strain 边
# 5. 写出 `mg_edgelist.txt`：两列，用 `\t` 分隔

# %% [code]
from itertools import combinations
from collections import defaultdict

def main():
    print(">>> Parsing GO OBO file...")
    parents, name_dict, namespace_dict, subset_dict = parse_go_obo(GO_OBO_FILE)

    print(">>> Building GO slim term set (generic + prok)...")
    slim_terms = build_slim_term_set(subset_dict, SLIM_SUBSETS)
    print(f"    # of slim terms: {len(slim_terms)}")

    # 所有边：统一存在一个 set 中，避免重复
    edges = set()

    # species -> list of strains（用于 strain-strain 边）
    species_to_strains = defaultdict(list)

    print(">>> Scanning GO annotation directories...")
    for strain_dir in sorted(os.listdir(GO_DIR)):
        strain_path = os.path.join(GO_DIR, strain_dir)
        if not os.path.isdir(strain_path):
            continue

        go_raw_file = os.path.join(strain_path, f"{strain_dir}.go_raw.tsv")
        if not os.path.exists(go_raw_file):
            print(f"    [WARN] GO raw file not found: {go_raw_file}")
            continue

        strain_name = strain_dir  # 直接用目录名作为 strain 节点名字
        species = parse_species_from_strain_dir(strain_dir)
        species_to_strains[species].append(strain_name)

        print(f"    Processing strain: {strain_name}  (species: {species})")

        # 读取 GO 原始注释
        df_go = pd.read_csv(go_raw_file, sep="\t", dtype=str).fillna("")
        if "GO_ID" not in df_go.columns:
            print(f"      [SKIP] No GO_ID column in: {go_raw_file}")
            continue

        # 去掉空 GO_ID
        df_go = df_go[df_go["GO_ID"].str.startswith("GO:")]
        if df_go.empty:
            continue

        # 对每个 strain 只保留 unique GO_ID 集合
        go_ids = sorted(df_go["GO_ID"].unique())

        for go_id in go_ids:
            # 向上 3 层找祖先（包含自身）
            anc = get_ancestors(go_id, parents, max_depth=UPWARD_DEPTH)

            # 在祖先中筛选 slim term
            slim_anc = [a for a in anc if a in slim_terms]

            if slim_anc:
                # 使用所有匹配到的 slim term
                target_go_ids = slim_anc
            else:
                # 若没有 slim 祖先，为了不丢失信息，退回原始 GO
                target_go_ids = [go_id]

            for tid in target_go_ids:
                go_label = format_go_label(tid, name_dict, namespace_dict)
                if not go_label:
                    continue
                # GO_term -> strain 边
                edges.add((go_label, strain_name))

    # 构建同 species 下的 strain-strain 边
    print(">>> Building strain-strain edges within the same species...")
    for species, strains in species_to_strains.items():
        if len(strains) < 2:
            continue
        for s1, s2 in combinations(sorted(set(strains)), 2):
            if s1 != s2:
                edges.add((s1, s2))

    # 输出 mg_edgelist.txt
    print(f">>> Writing mg_edgelist to: {MG_EDGE_FILE}")
    with open(MG_EDGE_FILE, "w", encoding="utf-8") as out_f:
        for src, dst in sorted(edges):
            out_f.write(f"{src}\t{dst}\n")

    print(f">>> Done. Total edges: {len(edges)}")

# 直接在 Notebook 里运行 main()
if __name__ == "__main__":
    main()
```

## Step 3. 准备 GGI_edgelists
为每个 GO annotation 准备单独的 GGI 网络
____

```python
from pathlib import Path
from collections import defaultdict
import pandas as pd
import re
from tqdm import tqdm

# ===== 路径配置 =====
DATA_ROOT = Path("/opt/ai4g_chriszyyang/buddy1/2_project_ongoing/4-antibio_resistance/PANACEA/data")

GO_ROOT   = DATA_ROOT / "4-go_annotation"
GGI_ROOT  = DATA_ROOT / "3-string_ppi_data"

# 版本1（无score）
OUT_ROOT1 = DATA_ROOT / "5-network" / "ggi_edgelists"
# 版本2（带score）
OUT_ROOT2 = DATA_ROOT / "5-network" / "ggi_edgelists_with_score"

OUT_ROOT1.mkdir(parents=True, exist_ok=True)
OUT_ROOT2.mkdir(parents=True, exist_ok=True)

print("GO_ROOT :", GO_ROOT)
print("GGI_ROOT:", GGI_ROOT)
print("OUT1:", OUT_ROOT1)
print("OUT2:", OUT_ROOT2)
```

```python
# ========== 第一步：读取 GO 注释 ==========

strain_gene_to_gos = dict()
all_go_annotations = set()

strain_dirs = [p for p in GO_ROOT.iterdir() if p.is_dir()]
print(f"Found {len(strain_dirs)} strain folders under GO_ROOT.")

for strain_dir in tqdm(strain_dirs, desc="Parsing GO annotations"):
    strain_name = strain_dir.name
    go_files = list(strain_dir.glob("*.go_raw.tsv"))
    if not go_files:
        continue

    go_file = go_files[0]
    df = pd.read_csv(go_file, sep="\t", dtype=str).fillna("")

    if not {"Genes", "GO_ID", "GO_Type"}.issubset(df.columns):
        print(f"[WARN] Missing columns in {go_file}, skip.")
        continue

    gene_to_gos = defaultdict(set)

    for _, row in df.iterrows():
        gene   = row["Genes"].strip()
        go_id  = row["GO_ID"].strip()
        go_typ = row["GO_Type"].strip()

        if not gene or not go_id or not go_typ:
            continue

        go_ann = f"{go_id}_{go_typ}"
        gene_to_gos[gene].add(go_ann)
        all_go_annotations.add(go_ann)

    if gene_to_gos:
        strain_gene_to_gos[strain_name] = gene_to_gos

print(f"Strains with GO annotations: {len(strain_gene_to_gos)}")
print(f"Total distinct GO annotations: {len(all_go_annotations)}")
```

```python
# ========== 第二步：读取 GGI，构建 GO-specific 子网（两种版本） ==========

# 无 score：go_ann → set((geneA, geneB))
go_to_edges_no_score = defaultdict(set)

# 带 score：go_ann → set((geneA, geneB, score))
go_to_edges_with_score = defaultdict(set)

def find_ggi_file_for_strain(strain_name: str):
    """
    寻找 <strain_name>.ggi.in_genes.tsv
    """
    subdir = GGI_ROOT / strain_name
    if subdir.is_dir():
        cand = subdir / f"{strain_name}.ggi.in_genes.tsv"
        if cand.exists():
            return cand
        files = list(subdir.glob("*.ggi.in_genes.tsv"))
        if files:
            return sorted(files)[0]

    # 根目录下
    cand2 = GGI_ROOT / f"{strain_name}.ggi.in_genes.tsv"
    if cand2.exists():
        return cand2

    # 再兜底
    files = [p for p in GGI_ROOT.glob("*.ggi.in_genes.tsv")
             if strain_name in p.name]
    if files:
        return sorted(files)[0]

    return None


for strain_name, gene_to_gos in tqdm(strain_gene_to_gos.items(),
                                    desc="Building GO-specific GGI subnets"):

    ggi_file = find_ggi_file_for_strain(strain_name)
    if ggi_file is None:
        print(f"[WARN] No .ggi.in_genes.tsv for {strain_name}")
        continue

    try:
        df_ggi = pd.read_csv(ggi_file, sep="\t", dtype=str).fillna("")
    except Exception as e:
        print(f"[ERROR] Reading {ggi_file}: {e}")
        continue

    if not {"gene1", "gene2", "score"}.issubset(df_ggi.columns):
        print(f"[WARN] gene1/gene2/score missing in {ggi_file}, skip.")
        continue

    for _, row in df_ggi.iterrows():
        g1 = row["gene1"].strip()
        g2 = row["gene2"].strip()
        score = row["score"].strip()

        if not g1 or not g2:
            continue

        gos1 = gene_to_gos.get(g1)
        gos2 = gene_to_gos.get(g2)
        if not gos1 or not gos2:
            continue

        # 找公共 GO 注释
        common = gos1 & gos2
        if not common:
            continue

        a, b = sorted((g1, g2))

        for go_ann in common:
            # 无 score 版
            go_to_edges_no_score[go_ann].add((a, b))
            # 带 score 版
            go_to_edges_with_score[go_ann].add((a, b, score))


print(f"GO with edges (no score): {len(go_to_edges_no_score)}")
print(f"GO with edges (with score): {len(go_to_edges_with_score)}")
```

```python
# ========== 第三步：输出两份 ggi_edgelists ==========
def safe_name(go_ann: str) -> str:
    s = go_ann.replace("/", "_")
    return re.sub(r"[^A-Za-z0-9._-]+", "_", s)

count1 = count2 = 0

for go_ann, edges in tqdm(go_to_edges_no_score.items(), desc="Writing NO-SCORE files"):
    if not edges:
        continue

    fname = safe_name(go_ann) + ".txt"
    out_path = OUT_ROOT1 / fname

    with out_path.open("w", encoding="utf-8") as f:
        for a, b in sorted(edges):
            f.write(f"{a} {b}\n")

    count1 += 1


for go_ann, edges in tqdm(go_to_edges_with_score.items(), desc="Writing SCORE files"):
    if not edges:
        continue

    fname = safe_name(go_ann) + ".txt"
    out_path = OUT_ROOT2 / fname

    with out_path.open("w", encoding="utf-8") as f:
        for a, b, score in sorted(edges):
            f.write(f"{a} {b} {score}\n")

    count2 += 1

print(f"Written {count1} GO edgelists (no score)")
print(f"Written {count2} GO edgelists (with score)")
```

```python
## 统计在不同 GO annotation 之间，比较它们的基因集合，并计算两两之间的交集
```

```python
from pathlib import Path
import pandas as pd

GGI_DIR = Path("/opt/ai4g_chriszyyang/buddy1/2_project_ongoing/4-antibio_resistance/PANACEA/data/5-network/ggi_edgelists")

go_to_genes = {}

for file in sorted(GGI_DIR.glob("*.txt")):
    go_ann = file.stem   # 文件名即 GO annotation

    genes = set()
    with file.open("r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 2:
                g1, g2 = parts[0], parts[1]
                genes.add(g1)
                genes.add(g2)

    go_to_genes[go_ann] = genes

print(f"Loaded {len(go_to_genes)} GO annotations with gene sets.")
```

```python
# 计算 GO 之间的两两基因交集（Intersection）
import pandas as pd
import itertools

gos = list(go_to_genes.keys())

intersection_matrix = pd.DataFrame(0, index=gos, columns=gos)

for g1, g2 in itertools.combinations(gos, 2):
    inter = len(go_to_genes[g1] & go_to_genes[g2])
    intersection_matrix.loc[g1, g2] = inter
    intersection_matrix.loc[g2, g1] = inter

# 对角线填充为各自基因数
for g in gos:
    intersection_matrix.loc[g, g] = len(go_to_genes[g])

intersection_matrix.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 12))
sns.heatmap(intersection_matrix, cmap="viridis")
plt.title("Gene Intersection Between GO Annotations")
plt.show()
```

```python
N = 30
subset = intersection_matrix.iloc[:N, :N]

plt.figure(figsize=(12, 10))
sns.heatmap(subset, cmap="viridis")
plt.title(f"Top {N} GO annotations (by gene intersection)")
plt.show()
```
