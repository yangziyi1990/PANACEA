```python
import os
import shutil
```

```python
# ====== 路径配置 ======
BASE = "/opt/ai4g_chriszyyang/buddy1/2_project_ongoing/4-antibio_resistance/PANACEA/data"
SRC_PPI = os.path.join(BASE, "3-string_ppi_data")
SRC_GO  = os.path.join(BASE, "4-go_annotation")

DEST_BASE = "/apdcephfs_qy3/share_2932069/ziyiyang/1-antibiotic_resistance/PANACEA/data"
DEST_PPI = os.path.join(DEST_BASE, "3-string_ppi_data")
DEST_GO  = os.path.join(DEST_BASE, "4-go_annotation")

# 目标目录不存在则创建
os.makedirs(DEST_PPI, exist_ok=True)
os.makedirs(DEST_GO, exist_ok=True)

# ====== 收集文件夹名 ======
ppi_dirs = {d for d in os.listdir(SRC_PPI) if os.path.isdir(os.path.join(SRC_PPI, d))}
go_dirs  = {d for d in os.listdir(SRC_GO) if os.path.isdir(os.path.join(SRC_GO, d))}

# ====== 求交集（共同命名的文件夹）======
common_dirs = ppi_dirs & go_dirs
print(f"共找到 {len(common_dirs)} 个共同命名文件夹")

# ====== 执行复制 ======
for d in sorted(common_dirs):
    src_ppi_d = os.path.join(SRC_PPI, d)
    src_go_d  = os.path.join(SRC_GO, d)

    dst_ppi_d = os.path.join(DEST_PPI, d)
    dst_go_d  = os.path.join(DEST_GO, d)

    print(f"复制: {d}")

    # 复制 PPI 文件夹
    if os.path.exists(dst_ppi_d):
        shutil.rmtree(dst_ppi_d)
    shutil.copytree(src_ppi_d, dst_ppi_d)

    # 复制 GO 文件夹
    if os.path.exists(dst_go_d):
        shutil.rmtree(dst_go_d)
    shutil.copytree(src_go_d, dst_go_d)

print("复制完成！")
```
